# Libraries

```python
import urllib.request
from bs4 import BeautifulSoup
import pandas as pd
import time
```

# Code

```python
def find_news(url):
    try:
        page=urllib.request.urlopen(url)
        soup = BeautifulSoup(page, 'html.parser')
        headline = soup.title.text
        #keywords = soup.find('meta', attrs={'name': 'news_keywords'})['content'].split(',')
        if soup.time == None:
            return (headline,'NA')
        else:
            return (headline,soup.time['datetime'])
    except:
        pass
```

```python
def handle_request(url):
    page = urllib.request.urlopen(url)
    request_soup = BeautifulSoup(page, 'html.parser')
    urls = [link['href'] for link in request_soup.find_all('a', attrs={'class':' fz-m'})]
    headline = []
    date = []
    for url in urls:
        result = find_news(url)
        #print(result)
        if result != None:
            headline.append(result[0])
            date.append(result[1])
    return pd.DataFrame({'headline':headline,'date':date})
```
